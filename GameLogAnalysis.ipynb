```python
def filterSixtySeconds(t):
    delta = datetime.timedelta(seconds=60)
#     t -= delta
#     if delta > datetime.timedelta(0):
#         t += datetime.timedelta(seconds=60)

    if t > delta:
        return True

    return 0
```

```python
final_table['valid_session'] = final_table.timeseconds.apply(filterSixtySeconds)
```
